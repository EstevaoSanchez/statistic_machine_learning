```python
# calculando o percentual de valores indefinidos em cada coluna numrica

def calcula_percentual_indefinido(dataframe, coluna ):
    indefinidos = dataframe.loc[pd.isna(dataframe[coluna]), coluna].shape[0]
    #total = dataframe.idade.shape[0]
    total = dataframe[coluna].shape[0]

    return (indefinidos/total)* 100
```

```python
'''
Aps a limpeza dos valores indefinidos, precisamos converter os
campos categricos em one-hot; so eles: classe, sexo e embarque. Para analis-los,
vamos comear obtendo os valores que podem assumir:
'''
df.classe_social.unique()
```

```python
'''
Isso quer dizer que um passageiro pode ter sido alocado na primeira, segunda ou
terceira classes. Computacionalmente esse tipo de categoria  muito ruim, porque 
comparvel! Sabemos que esses campos esto ordenados e no deveriam; ento vamos
mapear cada uma dessas classes para vetores. Para facilitar, vamos criar uma funo
que faa isso para ns:
'''

def one_hot(categorias, valor):
# Transforma as categorias num conjunto para eliminar duplicatas,
# depois transforma em lista para obter ndice
    categorias = list(set(categorias))
    print("Categorias: ")
    print(categorias)
# Obtm o ndice do elemento atual
    indice = categorias.index(valor)
    print("Indice ")
    print(indice)
# Cria um vetor com o nmero de coordenadas igual ao nmero de categorias
    print("Vetor")
    vetor = np.zeros(len(categorias))
# Modifica o ndice correspondente a classe para 1.0 no vetor de categorias
    vetor[indice] = 1.0
# devolve o vetor one-hot resultante
    return vetor
```

```python
'''
O prximo passo  normalizar os valores escalares. Vamos normalizar pela mdia, isto
, vamos aplicar a seguinte transformao:

  Xnovo = (Xatual - Xmdio)/ (Xmax - Xmin)

'''
def normaliza_media(valores):
    media = np.mean(valores)
    minimo = np.min(valores)
    maximo = np.max(valores)
    return (valores - media)/(maximo-minimo)
```

```python
df.idade = normaliza_media(df.idade)
df.head()
```

```python
df.parentes = normaliza_media(df.parentes)
df.head()
```

```python
df.dependentes = normaliza_media(df.dependentes)
df.dependentes.head(10)
```

```python
df.tarifa = normaliza_media(df.tarifa)
df.tarifa.head(10)
```

```python
'''
Aqui vamos criar um modelo baseado no conjunto de vizinhos mais prximos. Vamos
usar a biblioteca scikitlearn, ela possui algumas implementaes do algoritmo. O
primeiro passo  importar o que vamos precisar para o trabalho.
'''
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
#from bokeh.plotting import figure, show, output_notebook
```

```python
x = np.loadtxt("/home/entrada_treino.csv", delimiter=',')
y = np.loadtxt("/home/labels.csv", delimiter=',')
```

```python
y
```

```python
#cria um modelo para classificao usando KNN com k=5
model = KNeighborsClassifier(n_neighbors=5)
```

```python
'''
Com o modelo criado, o treinamento  simples, basta usar a funo fit referenciando os
vetores de entrada e os vetores com a classe selecionada. Vamos separar 70% dos dados
para treino e os outros 30% para validao. Primeiro, descobrimos quanto  70% dos
dados de treino:
'''
tamanho_treino = int(x.shape[0]*0.7)
tamanho_treino
```

```python
x.shape[0] #linhas
```

```python
x.shape[1] #colunas
```

```python
#Agora o que fazemos  separar para treino os dados at o ndice tamanho_treino:

x_t = x[:tamanho_treino]
y_t = y[:tamanho_treino]
```

```python
#Agora separamos a partir de tamanho_treino para validao
x_v = x[tamanho_treino:]
y_v = y[tamanho_treino:]
```

```python
#Agora criamos o modelo usando os dados de treino.
model.fit(x_t, y_t)
```

```python
'''
Vamos avaliar como o modelo performa nos dados de validao, ou seja, nos dados que
ele ainda no conhece. Calculando a acurcia do modelo
'''
model.score(x_v, y_v)
```
